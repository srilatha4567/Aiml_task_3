```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset
df = sns.load_dataset('iris')

# 1. Distribution of Numerical Features
plt.figure(figsize=(10, 6))
sns.histplot(df['sepal_length'], kde=True, color='skyblue')
plt.title('Distribution of Sepal Length')
plt.show()
# Insight: The distribution appears slightly right-skewed with a peak around 5.5cm.

# 2. Categorical Feature Analysis
plt.figure(figsize=(8, 5))
sns.countplot(x='species', data=df, palette='viridis')
plt.title('Count of Each Species')
plt.show()
# Insight: The dataset is perfectly balanced with 50 samples for each species.

# 3. Outlier Identification
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, palette='Set2')
plt.title('Box Plot for Outlier Detection')
plt.show()
# Insight: Sepal Width contains a few outliers above the upper whisker.

# 4. Correlation Heatmap
plt.figure(figsize=(8, 6))
# Dropping the categorical column for correlation
correlation_matrix = df.drop(columns=['species']).corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Feature Correlation Heatmap')
plt.show()
# Insight: Petal Length and Petal Width have a near-perfect positive correlation (0.96).

# 5. Identifying Important Features
sns.pairplot(df, hue='species', markers=["o", "s", "D"])
plt.show()
# Insight: Petal measurements are the most important features for distinguishing Setosa from other species.
```

EDA Defined: The process of using visual and statistical methods to "interview" data for patterns and anomalies before modeling

.Numerical Features: Histograms reveal skewness and data distribution (e.g., Bell Curve vs. Skewed)

.Categorical Features: Count plots identify class imbalance, showing if one category dominates the dataset.

Outlier Detection: Box plots highlight points outside the "normal" range that could distort a model's accuracy

.Correlation: A heatmap measures the strength ($r$ value) of relationships between features; high correlation ($r > 0.9$) indicates multicollinearity (redundancy)

.Importance of Visuals: They simplify complex data, making trends and "hidden" errors easy for stakeholders to see instantly

.Key Predictors: Features that show the most distinct separation between groups are prioritized for machine learning.
